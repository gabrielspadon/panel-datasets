```python
import os
import torch
import pandas as pd
```

```python
files_frame = []
for root, dirs, files in os.walk(".."):
    frames = []
    print(root)
    for filename in files:
        if filename[-3:] == "WTH":
            frames.append(pd.read_csv("%s/%s" % (root, filename), skiprows=4, sep='\s+', index_col=0))
    if len(frames) > 0:
        files_frame.append(pd.concat(frames, verify_integrity=True).sort_index())
```

```python
index = set(files_frame[0].index)
for df in files_frame:
    index = index & set(df.index)
```

```python
for i in range(len(files_frame)):
    files_frame[i] = files_frame[i].loc[index]
    files_frame[i] = files_frame[i].dropna(axis='columns')
```

```python
files_frame[0].index < 1000
```

```python
for i in range(len(files_frame)):
    files_frame[i] = files_frame[i].values
    files_frame[i] = torch.from_numpy(files_frame[i])
    files_frame[i] = torch.unsqueeze(files_frame[i], dim=0)
```

```python
final_tensor = torch.cat(files_frame)
```
